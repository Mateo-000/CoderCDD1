# **INTRODUCCION**

El conjunto de datos Ramen Ratings reúne más de 2.500 reseñas de fideos instantáneos registradas en “The Ramen Rater”, donde cada fila corresponde a una evaluación individual que contiene, entre otras variables, la marca, el nombre del producto (variety), el país de origen, el formato o estilo (cup, bowl, tray) y la calificación en una escala de 1 a 5 estrellas.

# **Hipotesis**

Este proyecto tiene como objetivo principal predecir la puntuación (stars) que un ramen recibirá a partir de atributos observables del envase y metadatos de la reseña. Desde ya podemos ver que este problema sera del tipo Regresion ya que intentaremos predecir valores numerico continuos.

Más allá del objetivo predictivo, interesa entender qué características (país, ingredientes implícitos en el nombre, estilo, marca) más influyen en la percepción de calidad y si existen patrones geográficos o de formulación que expliquen las diferencias de calificación.

**Hipótesis H1 — País y calidad**

Algunos países (ej. Japón, Corea, Singapur, etc.) tienden a tener medias de Stars diferentes de otros.

**Hipótesis H2 — Estilo (Cup / Bowl / Pack)**

El Style (Cup/Bowl/Pack/Tray) no afecta la puntuación.

# **MODELADO**

Como modelo voy a usar regresion lineal. Para poder prepararlo primero paso las variables strings a valores numericos, en el caso de la variedad tomo valores como el largo de la variedad y la cantidad de palabras que podrian asemejarce a la complejidad del plato.

En este caso voy a dividir el dataset en 80% para el entrenamiento y 20% para el testeo.

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# --- Recrear columnas numéricas sobre df ---
df['Variety'] = df['Variety'].astype(str)
df['Variety_len'] = df['Variety'].str.len()
df['Variety_words'] = df['Variety'].str.split().str.len()

brand_counts = df['Brand'].value_counts().to_dict()
df['Brand_count'] = df['Brand'].map(brand_counts)
country_counts = df['Country'].value_counts().to_dict()
df['Country_count'] = df['Country'].map(country_counts)

num_features = ['Variety_len', 'Variety_words', 'Brand_count', 'Country_count']
cat_features = ['Brand', 'Country', 'Style']

X = df[num_features + cat_features].copy()
y = df['Stars'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# --- Crear OneHotEncoder  ---
try:
    ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
except TypeError:
    ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)

# --- Pipeline / Preprocessor ---
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', ohe, cat_features)
    ],
    remainder='drop'
)

pipe = Pipeline([
    ('pre', preprocessor),
    ('model', Ridge(alpha=1.0, random_state=42))
])

# --- Cross-validation directa con sklearn ---
kf = KFold(n_splits=5, shuffle=True, random_state=42)

mse_cv = cross_val_score(pipe, X_train, y_train, scoring='neg_mean_squared_error', cv=kf, n_jobs=-1)
mae_cv = cross_val_score(pipe, X_train, y_train, scoring='neg_mean_absolute_error', cv=kf, n_jobs=-1)
r2_cv  = cross_val_score(pipe, X_train, y_train, scoring='r2', cv=kf, n_jobs=-1)
```

Luego de crear el pipeline con el proceso para el modelado y pasarle los valores que va a usar para entrenarse voy a ajustarlo y testearlo con el set que deje apartado para el test.

```python
# --- Ajuste final y evaluación en test ---
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# si ya tenés y_test y y_pred (como en tu celda anterior)
mse_test = mean_squared_error(y_test, y_pred)   # devuelve el MSE
rmse_test = np.sqrt(mse_test)                   # RMSE = sqrt(MSE)
mae_test  = mean_absolute_error(y_test, y_pred)
r2_test   = r2_score(y_test, y_pred)

print("Desempeño en test:")
print(f"RMSE: {rmse_test:.4f}")
print(f"MAE : {mae_test:.4f}")
print(f"R2  : {r2_test:.4f}")


# --- Matriz de confusión: discretizamos redondeando a enteros 1..5 ---
y_test_class = np.clip(np.rint(y_test), 1, 5).astype(int)
y_pred_class = np.clip(np.rint(y_pred), 1, 5).astype(int)

labels = [1,2,3,4,5]
cm = confusion_matrix(y_test_class, y_pred_class, labels=labels)

print("\nReporte de clasificación (Stars redondeadas):")
print(classification_report(y_test_class, y_pred_class, labels=labels, zero_division=0))

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels, cmap='Blues')
plt.xlabel('Predichas')
plt.ylabel('Reales')
plt.title('Matriz de confusión (Stars redondeadas)')
plt.show()
```

Claramente el modelo entrenado  y probado no es el mejor.

El R2 que habla sobre que tanto puede explicar la dispersion de la variable objetivo el modelo es muy bajo.

El accuracy esta por debajo del 42  esto significa que predice correctamente por debajo del 50% de las veces.

**Conclusion**:

 El modelo da predicciones aproximadas razonables (MAE < 1), pero no es confiable para predecir exactamente la cantidad de estrellas en muchos casos (accuracy 46%, R² muy bajo).

 A traves de la matriz de confusion podemos ver que hay un claro sesgo hacia la clase mayoritaria (4 estrellas) el modelo predice bien 4 porque es la clase más frecuente; en cambio falla en detectar extremos (1 y 5).

 Hay una clara falta de informacion, quiza con otros atributos podria haberse enriquecido mas el modelado.
